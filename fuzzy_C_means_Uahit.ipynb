```python
%matplotlib inline
import numpy as np
from fcmeans import FCM
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
n_samples = 5000

X = np.concatenate((
    np.random.normal((-2, -2), size=(n_samples, 2)),
    np.random.normal((2, 2), size=(n_samples, 2))
))
```

```python
X
```

```python
fcm = FCM(n_clusters=3)
fcm.fit(X)
```

```python
# outputs
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1], alpha=.1)
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels, alpha=.1)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=500, c='w')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 2)
```

```python
label = kmeans.fit_predict(X)
```

```python
#filter rows of original data
filtered_label0 = X[label == 0]
filtered_label1 = X[label == 1]
f, axes = plt.subplots(1, 2, figsize=(11,5))
#plotting the results
axes[0].scatter(X[:,0], X[:,1], alpha=1)
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1])
plt.show()
```

# Fuzy C Means

```python
X = x.to_numpy()
```

```python
fcm = FCM(n_clusters=3)
fcm.fit(X)
```

```python
# outputs
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1], alpha=.4)
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels, alpha=.4)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=150, c='w')
plt.show()
```

# **K - Means**

# k = 2

```python
kmeans = KMeans(n_clusters= 2)
```

```python
#filter rows of original data
filtered_label0 = X[label == 0]
filtered_label1 = X[label == 1]
f, axes = plt.subplots(1, 2, figsize=(11,5))
#plotting the results
axes[0].scatter(X[:,0], X[:,1], alpha=1)
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1])
plt.show()
```

# k = 3

```python
kmeans3 = KMeans(n_clusters= 3)
```

```python
labels = kmeans3.fit_predict(X)
```

```python
#filter rows of original data
filtered_label0 = X[labels == 0]
filtered_label1 = X[labels == 1]
filtered_label2 = X[labels == 2]
```

```python

f, axes = plt.subplots(1, 2, figsize=(11,5))
#plotting the results
axes[0].scatter(X[:,0], X[:,1], alpha=1)
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1])
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1])
plt.show()
```

# K-Means

```python
from numpy.core.fromnumeric import squeeze
original_target = squeeze(y.to_numpy())
original_target
```

```python
labels
```

```python
intersections = [i for i, j in zip(original_target, labels) if i == j]
accuracy = 100*len(intersections)/80
print("accuracy = ",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print(classification_report(original_target, labels))
```

```python
print(confusion_matrix(original_target, labels))
```
